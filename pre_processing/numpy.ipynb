```python
import numpy as np
```

## Reshape and Update In-Place

```python
e = np.arange(12)
print(e)
print("-------------")
print("f is a view of contents of e")
f = e.reshape(3, 4)
print(f)
print("Set values of e from index 5 onwards to 0")
e[5:] = 0
print(e)
print("f is also updated")
print(f)
print("-------------")
```

## Combine Arrays

```python
a = np.ones(3)
print(a)
print("---------")
b = np.array([2, 3, 4])
print(b)
print("----------------------------")
print(np.concatenate([a, b, a]))
print("------------------------------------------------------")
print("Use broadcasting when needed to do this automatically")
np.vstack([a, b, a])
print("------------------------------------------------------")
print("add new/concatenate features with hstack")
np.hstack([b, a])
```

## Create Sample Data

```python
%matplotlib inline

import pylab as plt
import seaborn

seaborn.set()
```

```python
# 1 row and 2 columns
plt.subplot(121)

# first plot

# Create evenly spaced numbers over the specified interval
x = np.linspace(0, 2, 10)

plt.plot(x, 'o-')
plt.title("A subplot with 1 line")
plt.xlabel('x lab')
plt.ylabel('y label')
# And now add something in the second part:
plt.subplot(122)

# Create sample data, add some noise
x1 = np.random.uniform(1, 100, 1000)
y = np.log(x1) + np.random.normal(0, .3, 1000)

plt.plot(x1, y, linestyle = 'none', marker = 'o', color = "orange", alpha = 0.3)
# change the breaks
plt.locator_params(axis = "x", nbins = 8)
plt.title("Scatterplot", loc = 'right')
plt.xlabel('x lab for this')
plt.ylabel('y is cool')

# Add a title:
plt.suptitle('A title common to my 2 plots', y = 1.02)

plt.show()
```
