```python
import pandas as pd
import numpy as np
from plotnine import *
%matplotlib inline
```

```python
(ggplot(df)
 + aes(
     x = 'albedo', 
     y = 'H', 
     color = 'neo')
 + geom_point()
 + labs(
     title = 'ggplot in python, woo', 
     x = 'albedo', 
     y = 'H')
)
```

```python
(ggplot(df)
 + aes(
     x = 'albedo', 
     y = 'H', 
     color = 'pha')
 + geom_point(alpha = 0.15)
 + labs(
     title = 'ggplot in python, woo', 
     x = 'albedo', 
     y = 'H')
)
```

```python
(ggplot(df)
 + aes(
     x = 'albedo', 
     y = 'H', 
     color = 'class',
     size = 'neo')
 + geom_point(alpha = 0.3)
 + labs(
     title = 'ggplot in python, woo', 
     x = 'albedo', 
     y = 'H')
)
```
